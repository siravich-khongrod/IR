### Remove the columns that contains no data

```
df.isnull().sum(axis=0)
```

```
nrow=df.shape[0]
NAcols = df.isnull().sum(axis=0).tolist() == np.repeat(nrow,df.shape[1])
print(df.columns[~NAcols])
df = df[df.columns[~NAcols]]
```

### By looking at the data drop other columns

```
print(df['condo.created_at'].isna().sum())
print(df['condo.created_at'].max())
print(df['condo.created_at'].min())
```

Since the range is only one year, this column is not very useful

```
unwantedCols = ['address.id', 'address.origin_id',
       'address.translate', 'address_id', 'agency_reference', 'conditions',
       'condo.created_at', 'condo.name',
       'condo.old_id',
       'condo.published_at',
       'condo.thumbnail_url', 'condo.updated_at',
       'condo.website', 'condo_id',
       'contract_advance', 'contract_deposit', 'contract_length', #most are NA
       'created_at',
       'detail.balcony_direction', #most are NA
       'detail.building',
       'detail.plan_picture_url', 'detail.property_broker',
       'detail.thumbnail_url', 'highlights','id',
       'post_description', 'post_title',
       'updated_at']
df = df[[c for c in df.columns if c not in unwantedCols]]
df.columns
```
